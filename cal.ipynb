```python
import numpy as np
import pandas as pd

file =('ArXiv_old.csv')
data= pd.read_csv(file)
```

```python
data.head(10)
```

```python
data.isna().any()
```

```python
data.dropna(inplace=True)
data.isna().any()
```

```python
data.head(10)
```

```python
print(data[['categories', 'created']].describe())
```

```python
import matplotlib.pyplot as plt
import pandas as pd

num = [0] * 16  # Initialize a list of length 16 (to represent the years from 1994 to 2009)
year = 1994

data['created'] = data['created'].astype(str).str[:4]  # Extract the first 4 characters and convert to string

for value in data['created']:
    if int(value) >= year and int(value) <= 2009:
        num[int(value) - year] += 1

for i in range(len(num)):
    print(f"Year {year + i}: {num[i]}")

years = list(range(1994, 2010))  # List of years from 1994 to 2009

plt.bar(years, num)
plt.title("Number of ArXiv articles per year")
plt.xlabel("Year")
plt.ylabel("number of articles")
plt.show()
```
