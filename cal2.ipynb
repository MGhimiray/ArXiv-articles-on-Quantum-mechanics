```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
num = [0] * 16
published=[0] *16
notp=[0]*16
year=1994

data['created'] = data['created'].astype(str).str[:4]  # Extract the first 4 characters and convert to string
data['doi'] = data['doi'].astype(str).str[:4]

for doi, value in zip(data['doi'], data['created']):
#print(data['created'])
#for doi in data['doi']:
 # for value in data['created']:#
    if int(value) >= year and int(value) <= 2009:
        num[int(value) - year] += 1
        if doi== '0.0':
          notp[int(value) - year]+=1
        else:
          published[int(value) - year]+=1

for i in range(len(num)):
    print(f"Year {year + i}: {num[i]}: {published[i]}: {notp[i]}")

years = list(range(1994, 2010))  # List of years from 1994 to 2009
bar_width = 0.46
bar_positions_published = np.arange(len(years))
bar_positions_notp = bar_positions_published + bar_width

# Plotting the bar diagrams side by side
plt.bar(bar_positions_published, published, width=bar_width, color='orange', label='Published')
plt.bar(bar_positions_notp, notp, width=bar_width, color='blue', label='Not Published')

# Set the x-axis tick labels
plt.xticks(bar_positions_published + bar_width / 2, years, rotation='vertical')

#plt.title("Number of ArXiv articles per year")
#plt.xticks(bar_positions_published + bar_width / 2, years)

plt.title("Published and unpublished articles in Arxiv")
plt.xlabel("Year")
plt.ylabel("Number of Articles")
plt.legend()
plt.show()

```
